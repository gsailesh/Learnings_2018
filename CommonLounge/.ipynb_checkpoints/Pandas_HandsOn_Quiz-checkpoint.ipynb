```python
import pandas as pd

diabetes = pd.read_csv('data/diabetes.csv')
diabetes.shape
```

```python
print diabetes[diabetes.Glucose == 0].shape
diabetes.Glucose = diabetes.Glucose.replace(0,None)
print diabetes[diabetes.Glucose == 0].shape
```

```python
cols = ['BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
print diabetes[cols].head(10)

for col in cols:
    print col, ": ", diabetes[diabetes[col] == 0].shape
    diabetes.loc[diabetes[col] == 0, col] = None
    print col, ": ", diabetes[diabetes[col] == 0].shape

print diabetes.info()
```

```python
diabetes.describe()
```

```python
diabetes.corr()
```

```python
import numpy as np

diabetes_normalized = diabetes
for col in diabetes_normalized.columns.values:
    col_mean = np.mean(diabetes[col])
    col_sd = np.std(diabetes[col])
    diabetes_normalized[col] = (diabetes_normalized[col] - col_mean)/col_sd

print diabetes_normalized.mean()
print diabetes_normalized.std()



# Or, in a single step,
# diabetes_normalized = (diabetes - diabetes.mean()) / diabetes.std()

diabetes_normalized.Outcome = diabetes_normalized.Outcome > 0.0
```

```python
diabetes_normalized.head(6)
```

```python
cols = [c for c in diabetes_normalized.columns.values if c not in ['Outcome']]

for col in cols:
    new_col = col+'_isnull'
    diabetes_normalized[new_col] = diabetes_normalized[col].fillna(1.0,inplace=True)
    diabetes_normalized.loc[diabetes_normalized[new_col] != 1.0,new_col] = 0.0
    
diabetes_normalized.shape
```
