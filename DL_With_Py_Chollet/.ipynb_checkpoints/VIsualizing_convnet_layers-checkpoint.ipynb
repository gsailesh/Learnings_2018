```python
model = VGG16(include_top=False)
model.summary()


LAYER_NAME = 'block1_conv1'
FILTER_INDEX = 0
```

```python
def tensor_to_image(x_tensor):
    
    x_tensor -= x_tensor.mean()
    x_tensor /= (x_tensor.std() + 1e-5)
    x_tensor *= 0.1

    x_tensor += 0.5
    x_tensor = np.clip(x_tensor,0,1)

    x_tensor *= 255
    x_tensor = np.clip(x_tensor,0,255).astype('uint8')

    return x_tensor

def generate_pattern(layer_name, filter_index, size=150):
    
    layer_output = model.get_layer(layer_name).output
    loss = K.mean(layer_output[:,:,:,filter_index])

    grads = K.gradients(loss, model.input)[0] # computes the gradient of input image w.r.t loss
    grads /= K.sqrt(K.mean(K.square(grads)) + 1e-5) # Gradient Normalization trick

    '''
    Above calculations for loss & grads are carried out below in `iterate`
    '''
    iterate = K.function(inputs=[model.input],outputs=[loss,grads]) # Returns gradient & loss for the given input

    input_image_data = np.random.random((1,size,size,3))*15 + 128 # Noisy

    step = 1.
    for iter in range(40): # Number of steps
        loss_val, grad_val = iterate([input_image_data])
        input_image_data += grad_val*step

    img = input_image_data[0]
    return tensor_to_image(img)
```

```python
plt.imshow(generate_pattern(LAYER_NAME,FILTER_INDEX)) # shows Filter activations
```

```python
size = 64
margin = 3
LAYER_NAME = 'block3_conv1'

results = np.zeros((8*size + 7*margin, 8*size + 7*margin, 3))

for i in range(8):
    for j in range(8):
        filter_num = i+(j*8)
        filter_img = generate_pattern(LAYER_NAME, filter_num, size=size)
        
        h_start = i*size + i*margin
        h_end = h_start + size
        
        v_start = j*size + j*margin
        v_end = v_start + size
        
#         print("Filter: ", filter_num, "Coords: ", h_start,h_end,v_start,v_end)
        results[h_start:h_end, v_start:v_end, :] = filter_img

plt.figure(figsize=(20,20))
plt.imshow(results)
```
